```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Media variables 2021-11-18.xlsx')
```

```python
df.head()
```

```python
df_train = df[df['Date']<='2021-06-28']
df_test = df[df['Date']>'2021-06-28']
```

```python
y = np.array(df_train['FAA Lead Form Impressions'])
y
```

```python
X = np.array(df_train[['Social-Brand','Social-GO','Social-Lumino','Social-Sponsorship','Social-Other']])
X
```

```python
#Decay effect as in MassTer
def decay(grp, r):
    adstock = np.zeros(len(grp))
    adstock[0] = grp[0]
    for i in range(1,len(grp)):
        adstock[i] = grp[i] + r*adstock[i-1]
    return adstock
```

```python
test = df_train['Social-Brand']
dsocial = decay(test,.8)
fig = plt.figure(figsize = (12,5))
plt.plot(df_train['Date'],test,'r-',label = 'Social-Brand')
plt.plot(df_train['Date'],dsocial,'b-.',label = 'Social-Brand with Decay')
plt.legend()
plt.title('Decay Effect')
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.show()
```

```python
#Split into train and val
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse

def rmse(actual,pred):
    return np.sqrt(mse(actual,pred))

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)

X_val.shape
```

```python
#Pick the best decay effect for each social variable, initialize population
np.random.seed(0)
pop_size = 20
initial_pop = np.random.rand(pop_size,X_train.shape[1])
initial_pop
```

```python
def calculate_fitness(individual):
    tmedia_decay = np.zeros(X_train.shape)
    vmedia_decay = np.zeros(X_val.shape)
    for i, r in enumerate(individual):
        tmedia_decay[:,i] = decay(X_train[:,i],r)
        vmedia_decay[:,i] = decay(X_val[:,i],r)

    scaler.fit(tmedia_decay)
    tscaled = scaler.transform(tmedia_decay)
    vscaled = scaler.transform(vmedia_decay)

    lin_reg = LinearRegression()
    lin_reg.fit(tscaled,y_train)
    
    tpred = lin_reg.predict(tscaled)
    vpred = lin_reg.predict(vscaled)
    
    return rmse(y_val,vpred)
```

```python
def select_parents(population,tournament_size,pop_size):
    rmse_log = np.zeros(pop_size)
    for i, ind in enumerate(population):
        rmse_log[i] = calculate_fitness(ind)
    
    #Use tournament selection to get next generation
    parents = np.zeros(population.shape)
    for i in range(0,pop_size):
        players = np.random.randint(0,pop_size,tournament_size)
        rmse_players = rmse_log[players]
        fittest_index = int(np.where(rmse_log == min(rmse_players))[0])
        parents[i] = population[fittest_index,:]

    return parents
```

```python
select_parents(initial_pop,5,20)
```
